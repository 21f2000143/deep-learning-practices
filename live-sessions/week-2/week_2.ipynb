```python
sentences = [
  "The quick brown fox jumps over the lazy dog.",
  "Artificial intelligence is transforming the world.",
  "I love learning about natural language processing.",
  "Transformers are powerful models for text data.",
  "Python is a versatile programming language.",
  "Deep learning enables amazing applications.",
  "Data science combines statistics and computing.",
  "Machine learning models require lots of data.",
  "Tokenization is the first step in NLP pipelines.",
  "OpenAI develops advanced AI technologies."
]
```

```python
from tokenizers import Tokenizer, models, pre_tokenizers, trainers
```

```python
tokenizer = Tokenizer(models.BPE(unk_token='[UNK]'))
```

```python
from tokenizers.normalizers import Lowercase
from tokenizers.pre_tokenizers import Whitespace
```

```python
tokenizer.normalizer = Lowercase()
tokenizer.pre_tokenizer = Whitespace()
```

```python
trainer = trainers.BpeTrainer(
  vocab_size = 50,
  spcial_tokens = ['[UNK]', '[CLS]', '[SEP]', '[MASK]'],
  continuing_subword_prefix = '##'
)
```

```python
from tokenizers.processors import TemplateProcessing

tokenizer.post_processor = TemplateProcessing(single='[CLS] $0 [SEP]',
                                              pair = '[CLS] $A [SEP] $B',
                                              special_tokens=[('[CLS]', 2), ('[SEP]', 3), ('[UNK]', 4)])
```

```python
tokenizer.train_from_iterator(sentences, trainer)
```

```python
tokenizer.save('cusetom_bpe_tokenizer.json')
```

```python
bpe_tokenizer = Tokenizer.from_file('cusetom_bpe_tokenizer.json')
```

```python
input_test = "Artificial intelligence is transforming the world."
encoded = bpe_tokenizer.encode(input_test)
pprint(encoded)
```
